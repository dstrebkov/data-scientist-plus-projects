Значение метрики `RMSE` на тестовой выборке должно быть не больше `48`.

## Описание данных

Данные лежат в файле /datasets/taxi.csv. Скачать датасет. 
Количество заказов находится в столбце 'num_orders' (от англ. number of orders, «число заказов»).

```python
import datetime

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import seasonal_decompose

from sklearn.model_selection import train_test_split, cross_val_score, TimeSeriesSplit, RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

from lightgbm import LGBMRegressor
```

### Шаг 1. Загрузите данные.

```python
data = pd.read_csv('...', index_col=[0], parse_dates=[0])
display(data.head())
print('Информация по набору данных:\n')
print(data.info())
print()
print(f'Размерность набора данных: {data.shape}')
```

### Шаг 2. Выполните их ресемплирование по одному часу.

Перед выполнением ресемплирования проверим монотонность нашего индекса:

```python
print(f'Индекс монотонный? {("Да" if data.index.is_monotonic else "Нет")}')
```

Индекс оказался монотонным, поэтому выполнять `sort_index()` не требуется, можно переходить к ресемплированию.

Используем `sum()`, раз нам потребуется спрогнозировать количество заказов на следующий час:

```python
data = data.resample('1H').sum()
```

```python
display(data)
```

Данных стало ожидаемо меньше, `4416` записей. 

Теперь для более полной визуализации добавим новый столбец со скользящим средним по `24` часам в наш набора данных:

```python
data['rolling_mean_for_plot'] = data.rolling(24).mean()
```

Взглянем на график:

```python
def get_x_labels(date_min, date_max, freq):
    labels = []
    for x in pd.date_range(start=date_min, end=date_max, freq=freq):
        labels.append(x.strftime("%Y-%m-%d %H:%M:%S"))
    return labels

fig, ax = plt.subplots(figsize=(16, 8))
_ = sns.lineplot(x=data.index, y=data['num_orders'], ax=ax)
_ = sns.lineplot(x=data.index, y=data['rolling_mean_for_plot'], label='Скользящее среднее по 24 часам', ax=ax)
_ = ax.set_xlabel('Дата')
_ = ax.set_ylabel('Количество заказов')
_ = ax.set_title('Распределение количества заказов по датам (часы)')
_ = ax.tick_params(axis='x', rotation=60)
_ = ax.set_xticks(get_x_labels(np.min(data.index), np.max(data.index) + datetime.timedelta(days=15), freq='15D'))
_ = ax.legend(loc='upper left')
_ = ax.grid(True)
```

Видя поведение скользящего среднего, можно заметить некоторую периодичность в подъёмах и спусках, а также общее стремлению к увеличению значения ближе к концу графика. Но дать более точную интерпретацию мне сложновато.

Может быть получится получить чуть больше информации, выделив компоненты _тренда_ и _сезонности_ в нашем наборе данных, и посмотрим на их графическое представление:

```python
data = data.drop(['rolling_mean_for_plot'], axis=1) # заодно удалим созданный столбец со скользящим средним
decomposed = seasonal_decompose(data)

fig, ax = plt.subplots(figsize=(22, 6))
decomposed.trend.plot(ax=ax, color='blue', label='Тренд')
decomposed.seasonal.plot(ax=ax, color='orange', label='Сезонность')
_ = ax.legend(loc='best')
_ = ax.grid(True)
```

Можно сказать, что график тренда показывает увеличение количества заказов такси с течением времени.  Что касается графика сезонности, то он не явлется наглядным, поэтому показать сезонность в данных по месяцам, наверное, он не способен - нужно более значительное укрупнение, т.е. рассмотрение менее длительных промежутков времени, в течение которых можно было бы опять попробовать определить сезонность. Сделаем это:

```python
N_PLOTS = 3

for i in range(N_PLOTS):
    fig, ax = plt.subplots(figsize=(22, 2))
    date_min = datetime.datetime.strptime(f'2018-{i + 3}-15', '%Y-%m-%d')
    date_max = datetime.datetime.strptime(f'2018-{i + 3}-17', '%Y-%m-%d')
    seasonal_slice = decomposed.seasonal[date_min:date_max]
    seasonal_slice.plot(ax=ax, color='orange', label='Сезонность')
    _ = ax.grid()
    _ = ax.set_title(f'График сезонности за 2 дня в середине месяца ({date_min.strftime("%B")})')
    _ = ax.set_xlabel('Время')
    fig.autofmt_xdate()
```

На основе графиков сделаем несколько выводов:

- на мой взгляд, сезонность наблюдается в пределах дня - в том смысле, что количество заказов в течение суток следует определённому паттерну, причём глядя на данные за несколько месяцев, можно убедиться, что эта особенность на самом деле имеет место;

- пиками по количеству заказов в течение дня являются времена около `00:00` (когда нет возможности воспользоваться общественным транспортом, в районе `09:00` (когда люди добираются на работу, если это будний день) и в районе `17:00`, когда начинается перевозка людей, закончивших свою работу;

- явный спад наблюдается в районе `06:00`, видимо, в это время уже закончилась ночная перевозка пассажиров, и но ещё не началась утренняя.

### 4. Добавление новых признаков



Прежде чем создавать тренировочную и тестовую выборки, добавим новые признаки в нашу модель:

```python
def make_features(data, max_lag, rolling_mean_size, rolling_mean_counts=3):
    data['day'] = data.index.day
    data['month'] = data.index.month
    data['day_of_week'] = data.index.dayofweek
    data['weekend_or_not'] = data.index.day_name().isin(['Saturday', 'Sunday']).astype(int)
    data['quarter'] = data.index.quarter;
    data['seconds'] = data.index.view(int)
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    for rolling_mean in range(rolling_mean_size - rolling_mean_counts, rolling_mean_size + 1):
        data[f'rolling_mean_{rolling_mean}'] = data['num_orders'].rolling(rolling_mean, closed='left').mean()
        
    data = data.dropna()
    return data

data = make_features(data, 20, 20)

display(data)
```

### Шаг 5. Сделайте тестовую выборку размером `10%` от исходных данных. Обучите разные модели с различными гиперпараметрами.

```python
Xy_train, Xy_test = train_test_split(data, shuffle=False, test_size=0.1)
print(data.shape, Xy_train.shape, Xy_test.shape)

y_train = Xy_train['num_orders']
y_test  = Xy_test['num_orders']
X_train = Xy_train.drop(['num_orders'], axis=1)
X_test  = Xy_test.drop(['num_orders'], axis=1)

print()
print(f'X_train.shape = {X_train.shape}')
print(f'X_test.shape  = {X_test.shape}')
print(f'y_train.shape = {y_train.shape}')
print(f'y_test.shape  = {y_test.shape}')
```

Поскольку у нас нет валидационной выборки, то оценивать качество будем на основе кросс-валидации.

Из-за того, что мы работаем с временными рядами, воспользуемся объектом `TimeSeriesSplit` для создания корректного кросс-валидационного разбиения ряда:

```python
cv = TimeSeriesSplit(n_splits=10)
```

Теперь обучим несколько моделей в версии по умолчанию, т.е. без подбора наилучших гиперпараметров:

```python
rmse_linear_reg =\
    -np.mean(cross_val_score(LinearRegression(), X_train, y_train, cv=cv, scoring='neg_root_mean_squared_error'))
rmse_decision_tree =\
    -np.mean(cross_val_score(DecisionTreeRegressor(random_state=0), X_train, y_train, cv=cv, scoring='neg_root_mean_squared_error'))
rmse_random_forest =\
    -np.mean(cross_val_score(RandomForestRegressor(random_state=0), X_train, y_train, cv=cv, scoring='neg_root_mean_squared_error'))
rmse_lgbm =\
    -np.mean(cross_val_score(LGBMRegressor(random_state=0), X_train, y_train, cv=cv, scoring='neg_root_mean_squared_error'))

print(f'RMSE линейной регрессии : {rmse_linear_reg:.3f}')
print(f'RMSE дерева решений     : {rmse_decision_tree:.3f}')
print(f'RMSE случайного леса    : {rmse_random_forest:.3f}')
print(f'RMSE LGBMRegressor      : {rmse_lgbm:.3f}')
```

Теперь попробуем подобрать наилучшие гиперпараметры для дерева решений, случайного леса и `LGBMRegressor`:

```python
# Здесь выставлено значение N_ITER = 1
# В действительности этому предшествовало выставление N_ITER = 500 и поиск наилучшей комбинации на 500 итерациях
# Для каждой модели ниже выставлены единственные (наилучшие) значения гиперпараметров
N_ITER = 1 # 500

def search_best_hyperparams(model, X, y, params, jobs=-1):
    randomized_search = RandomizedSearchCV(estimator=model, param_distributions=params, scoring='neg_root_mean_squared_error', cv=10, n_iter=N_ITER, random_state=0)
    randomized_search = randomized_search.fit(X, y)
    best_score = randomized_search.best_score_
    best_params = randomized_search.best_params_
    print(f'Hyperparameter tuning for "{model.__class__.__name__}":\n\tBest RMSE on Train = {-best_score}\n\tBest params = {best_params}')
    return -best_score, best_params
```

Сначала подберём наилучшие гиперпараметры для дерева решений:

```python
decision_tree_params = {
    'max_features': ['auto'],      # 'max_features'      : ['auto', 'sqrt'],
    'max_depth': [10],             # 'max_depth'         : [3, 10, 15, 20, 30, 50, 75, 150],
    'min_samples_split': [30],     # 'min_samples_split' : [2, 5, 10, 15, 30, 50, 75, 100, 125, 150],
    'min_samples_leaf': [30],      # 'min_samples_leaf'  : [1, 2, 4, 8, 10, 15, 20, 25, 30, 50, 75],
    'criterion': ['squared_error'] # 'criterion'         : ['squared_error', 'absolute_error']
}

tree_best_score, tree_best_params = search_best_hyperparams(DecisionTreeRegressor(random_state=0), X_train, y_train, decision_tree_params)
tuned_tree = DecisionTreeRegressor(random_state=0, max_features=tree_best_params['max_features'], max_depth=tree_best_params['max_depth'],
                                    min_samples_split=tree_best_params['min_samples_split'], min_samples_leaf=tree_best_params['min_samples_leaf'],
                                    criterion=tree_best_params['criterion'])
```

Теперь сделаем то же самое для случайного леса:

```python
random_forest_params = {
    'n_estimators': [700],    # 'n_estimators'      : [50, 75, 100, 125, 150, 200, 250, 300, 400, 500, 600, 700, 800, 900],
    'max_features': ['sqrt'], # 'max_features'      : ['auto', 'sqrt'], 
    'max_depth': [50],        # 'max_depth'         : [3, 10, 20, 50, 100, 150, 175, 200, 225, 250, 300, 350, 400],
    'min_samples_split': [2], # 'min_samples_split' : [2, 5, 10, 15, 30, 50, 100, 125, 150, 175, 200],
    'min_samples_leaf': [2],  # 'min_samples_leaf'  : [2, 4, 8, 10, 15, 30, 50, 75, 100, 125, 150],
    'bootstrap': [False]      # 'bootstrap'         : [True, False]
}

forest_best_score, forest_best_params = search_best_hyperparams(RandomForestRegressor(random_state=0, n_jobs=-1), X_train, y_train, random_forest_params, jobs=None)
tuned_forest = RandomForestRegressor(random_state=0, n_jobs=-1, n_estimators=forest_best_params['n_estimators'], max_features=forest_best_params['max_features'],
                                     max_depth=forest_best_params['max_depth'], min_samples_split=forest_best_params['min_samples_split'],
                                     min_samples_leaf=forest_best_params['min_samples_leaf'], bootstrap=forest_best_params['bootstrap'])
```

И, наконец, для градиентного бустинга `LGBMRegressor`:

```python
lgbm_params = {
    'num_leaves': [31],      # 'num_leaves'    : [3, 5, 7, 14, 21, 28, 31, 50, 75, 100, 150],
    'learning_rate': [0.03], # 'learning_rate' : [0.001, 0.01, 0.1, 0.03, 0.003, 0.03, 0.1, 0.15, 0.25, 0.3],
    'max_depth': [32],       # 'max_depth'     : [1, 2, 3, 5, 7, 9, 12, 14, 16, 18, 20, 22, 24, 26, 32, 34, 36],
    'n_estimators': [650]    # 'n_estimators'  : [25, 50, 75, 100, 125, 150, 200, 300, 500, 650, 700, 800, 850, 900],
}

lgbm_best_score, lgbm_best_params = search_best_hyperparams(LGBMRegressor(random_state=0, n_jobs=-1), X_train, y_train, lgbm_params)
tuned_lgbm = LGBMRegressor(random_state=0, n_jobs=-1, n_estimators=lgbm_best_params['n_estimators'],
                           num_leaves=lgbm_best_params['num_leaves'], learning_rate=lgbm_best_params['learning_rate'],
                           max_depth=lgbm_best_params['max_depth'])
```

Оформим предварительные выводы в виде таблицы:

```python
interm_results_df = pd.DataFrame(
    {
        'Model': [
            LinearRegression().__class__.__name__,
            DecisionTreeRegressor().__class__.__name__,
            RandomForestRegressor().__class__.__name__,
            LGBMRegressor().__class__.__name__
        ],
        'RMSE (модель по умолчанию)': [
            f'{rmse_linear_reg:.3f}',
            f'{rmse_decision_tree:.3f}',
            f'{rmse_random_forest:.3f}',
            f'{rmse_lgbm:.3f}'
        ],
        'RMSE (улучшенная модель)': [
            'N/A',
            f'{tree_best_score:.3f}',
            f'{forest_best_score:.3f}',
            f'{lgbm_best_score:.3f}'
        ]
    }
).set_index('Model')

display(interm_results_df)
```

### Шаг 6. Проверьте данные на тестовой выборке

Наилучшим с точки зрения RMSE образом себя показала модель `LGBMRegressor`, поэтому именно её и будем финально прогонять на тестовой выборке ниже:

```python
best_model = LGBMRegressor(**lgbm_best_params)

best_model = best_model.fit(X_train, y_train)
y_pred_test = best_model.predict(X_test)

rmse_test = mean_squared_error(y_test, y_pred_test, squared=False)

print(f'Final model "{best_model.__class__.__name__}"')
print(f'\tRMSE на тестовой выборке: {rmse_test}')
```

### Шаг 7. Сделайте выводы.

Прежде чем сформулировать итоговый вывод, попробуем взглянуть на важность признаков обученной в итоге модели градиентного бустинга:

```python
def plot_boosting_feature_importances(model, feature_names):
    feature_importances = (model.feature_importances_ / sum(model.feature_importances_)) * 100.
    results = pd.DataFrame({'Features': feature_names, 'Importances': feature_importances})
    results.sort_values(by='Importances', inplace=True)

    fig, ax = plt.subplots(figsize=(12, 6))
    _ = ax.barh(results['Features'], results['Importances'])
    _ = ax.set_title(f'Столбчатая диаграмма важности признаков модели {model.__class__.__name__}')
    _ = ax.set_xlabel('Важность (%)')
    _ = ax.grid(True)
    
plot_boosting_feature_importances(best_model, X_train.columns.values)
```

В ходе проекта была построена модель для предсказания количества заказов такси на следующий час.

Помимо признаков, связанных со скользящим средним и смещением в набор данных были добавлены дополнительные признаки:

- является ли текущий день выходным или рабочим (`weekend_or_not`);

- квартал для текущего дня (`quarter`);

Как можно видеть на диаграмме важности признаков, именно `lag`-признаки так ценны для качества модели. Также можно отметить добавленный признак `seconds`, который занял `4` место по _пользе_ для построенной в итоге модели.

В целом же в процессе работы над проектом рассматривались несколько моделей, способных решить такую задачу регрессии временного ряда:

- линейная регрессия;

- дерево решений для регрессии;

Именно последняя модель и показала наилучшее качество при кросс-валидации на тренировочном наборе, поэтому впоследствии она же и была обучена на тестовой выборке, продемонстрировав итоговое качество в `44.57` заказов по RMSE.
