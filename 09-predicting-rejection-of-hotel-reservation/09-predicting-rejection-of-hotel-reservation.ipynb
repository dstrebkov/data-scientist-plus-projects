# Описание проекта

Заказчик этого исследования — сеть отелей «*Как в гостях*».
Чтобы привлечь клиентов, эта сеть отелей добавила на свой сайт возможность забронировать номер без предоплаты. Однако если клиент отменял бронирование, то компания терпела убытки. Сотрудники отеля могли, например, закупить продукты к приезду гостя или просто не успеть найти другого клиента.

Чтобы решить эту проблему, вам нужно разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — `80%` от стоимости номера **за одни сутки** и затрат **на разовую уборку**. Деньги будут списаны со счёта клиента, если он всё же отменит бронь.

## Бизнес-метрика и другие данные

Основная бизнес-метрика для любой сети отелей — её **прибыль**. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.
В отеле есть несколько типов номеров. В зависимости от типа номера назначается стоимость за одну ночь. Есть также затраты на уборку. Если клиент снял номер надолго, то убираются **каждые два дня**.

#### Стоимость номеров отеля:
- категория `A`: за ночь — `1 000`, разовое обслуживание — `400`;
- категория `B`: за ночь — `800`, разовое обслуживание — `350`;
- категория `C`: за ночь — `600`, разовое обслуживание — `350`;
- категория `D`: за ночь — `550`, разовое обслуживание — `150`;
- категория `E`: за ночь — `500`, разовое обслуживание — `150`;
- категория `F`: за ночь — `450`, разовое обслуживание — `150`;
- категория `G`: за ночь — `350`, разовое обслуживание — `150`.

В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на `20%`, летом — на `40%`.

На разработку системы прогнозирования заложен бюджет — `400 000`. При этом необходимо учесть, что внедрение модели должно окупиться **за год**. Затраты на разработку должны быть меньше той выручки, которую система принесёт компании.

Сделаем описанные выше изменения в отдельной функции `preprocess_columns()`:

````python
PRICE_PER_NIGHT_AND_SERVICE = {
    'single_night': {
        'A': 1000,
        'B': 800,
        'C': 600,
        'D': 550,
        'E': 500,
        'F': 450,
        'G': 350
    },
    
    'single_service': {
        'A': 400,
        'B': 350,
        'C': 350,
        'D': 150,
        'E': 150,
        'F': 150,
        'G': 150
    }
}

def preprocess_columns(df):
    df = df.assign(
        adults = lambda x: x['adults'].astype(int),
        babies = lambda x: x['babies'].astype(int),
        children = lambda x: x['children'].astype(int),
        lead_time = lambda x: x['lead_time'].astype(int),
        
        reserved_room_type = lambda x: x['reserved_room_type'].str.strip(),
        meal = lambda x: x['meal'].str.strip()
    )
   
    df['per_night_price'] = df['reserved_room_type'].apply(
        lambda room_type: PRICE_PER_NIGHT_AND_SERVICE['single_night'][room_type])

    df['per_service_price'] = df['reserved_room_type'].apply(
        lambda room_type: PRICE_PER_NIGHT_AND_SERVICE['single_service'][room_type])

    return df

df_train = preprocess_columns(df_train)
df_test = preprocess_columns(df_test)

display(df_test.tail())
````

Далее выполним удаление выбросов из упомянутых выше признаков. Под выбросом будем понимать **любое значение, которое на 1.5 IQR меньше первого квартиля или на 1.5 IQR больше третьего квартиля**:

````python
outlier_cleaning_cols = [
    'lead_time',
    'previous_cancellations',
    'previous_bookings_not_canceled',
    'booking_changes',
    'days_in_waiting_list',
]

def remove_outliers(data):
    outliers_statistics = dict()

    for column in outlier_cleaning_cols:
        # Получаем первый и третий квартили
        q1 = data[column].quantile(0.25)
        q3 = data[column].quantile(0.75)

        # Получаем межквартильный размах
        iqr = q3 - q1

        # Подсчитываем верхнюю и нижнюю границы, все объявления вне которых будем считать выбросами
        lower_bound = max(0, q1 - 1.5 * iqr)
        upper_bound = q3 + 1.5 * iqr
        
        outliers_frame = data.query(f'{column} < @lower_bound or {column} > @upper_bound')
        
        outliers_statistics[column] = outliers_frame.shape[0]

        # Удаляем значения вне интервала от "Q1 - 1.5*IQR" до "Q3 + 1.5*IQR"
        data.drop(outliers_frame.index, inplace=True)

    # Выведем статистику по удалённым выбросам
    for k, v in sorted(outliers_statistics.items(), key=lambda kv: kv[0]):
        print(f'    {k} : {v}')

print('Количество удалённых выбросов по категориям для набора "df_train":')
remove_outliers(df_train)
print()
print('Количество удалённых выбросов по категориям для набора "df_test":')
remove_outliers(df_test)
````

### Шаг 3. Вычислите бизнес-метрику

#### - Оцените прибыль отеля без внедрения депозитов.

Начнём с того, что вновь расширим наши наборых данных, в этот раз для того, чтобы учесть сезонность:

        В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.
        
Для учёта данных особенностей создадим вспомогательный словарь `SEASONS_DICT`, который будет способен по индексу месяца возвращать имя месяца (просто на всякий случай), а также информацию о сезоне года:

- индекс сезона (начиная с `0`);
- множитель, который будет использован для учёта тех самых сезонных коэффициентов в `20%` и `40%`, которые были упомянуты выше.

````python
SEASONS_DICT = {
    # Зима
    1:  { 'name': 'January',   'season_id': 0, 'multiplier': 1.0 },
    2:  { 'name': 'February',  'season_id': 0, 'multiplier': 1.0 },
    
    # Весна
    3:  { 'name': 'March',     'season_id': 1, 'multiplier': 1.2 },
    4:  { 'name': 'April',     'season_id': 1, 'multiplier': 1.2 },
    5:  { 'name': 'May',       'season_id': 1, 'multiplier': 1.2 },
    
    # Лето
    6:  { 'name': 'June',      'season_id': 2, 'multiplier': 1.4 },
    7:  { 'name': 'July',      'season_id': 2, 'multiplier': 1.4 },
    8:  { 'name': 'August',    'season_id': 2, 'multiplier': 1.4 },

    # Осень 
    9:  { 'name': 'September', 'season_id': 3, 'multiplier': 1.2 },
    10: { 'name': 'October',   'season_id': 3, 'multiplier': 1.2 },
    11: { 'name': 'November',  'season_id': 3, 'multiplier': 1.2 },

    # Зима
    12: { 'name': 'December',  'season_id': 0, 'multiplier': 1.0 },
}
````

Далее с помощью функции `set_in_out_dates()` создадим дополнительные столбцы, содержащие точные даты заезда (`in_date`) и отъезда (`out_date`) для каждого из сделанных бронирований; применим функцию для тренировочной и тестовой выборок:

````python
def set_in_out_dates(df):
    # Выставляем отдельным столбцом дату прибытия
    df['in_date'] = df.apply(lambda r:
        pd.to_datetime(f"{r['arrival_date_year']}-{r['arrival_date_month']}-{r['arrival_date_day_of_month']}"),
        axis=1)
    # Выставляем отдельным столбцом дату отъезда
    df['out_date'] = df.apply(lambda r:
        r['in_date'] + datetime.timedelta(days=r['total_nights']),
        axis=1)

set_in_out_dates(df_train)
set_in_out_dates(df_test)

df_test.sample(n=5, random_state=1)
````

Теперь оценим максимальную длительность бронирований. Это нужно для того, чтобы оценить возможный случай, когда одно бронирование затрагивает `3` и более сезонов года (т.е. случаи, когда происходит два пересечения сезонов и более в рамках одного бронирования):

Видим, что максимальные бронирования не очень длинные - всего `10` ночей, т.е. возможные пересечения сезонов сводятся к тому, что может наблюдаться переход из одного сезона в другой в рамках одного бронирования (с лета на осень, с зимы на весну и т.д.).

Наконец, напишем функцию для **подсчёта прибыли**, применять её будем на тестовой выборке. Здесь нужно отметить, что написанная функция должна быть способна обрабатывать прибыль как до внедрения депозитов, так и после. Во втором случае в функции нужно будет учесть состояние столбца `is_canceled_predicted`, который появится в нашем проекте позднее - после создания модели машинного обучения, способной делать предсказания отказов. Сейчас же, пока такого столбца нет, функция должна отрабатывать не принимая во внимание возможность его присутствия:

````python
def get_number_of_services_by_nights_count(nights_count):
    '''Возвращает количество уборок, которое должно быть проведено в номере, забронированном
    на "nights_count" количество ночей согласно правилу:
            "Если клиент снял номер надолго, то убираются каждые два дня"
    '''
    return math.ceil(nights_count / 2)

def calculate_partial_income(nights, season, per_night_price, per_service_price):
    '''Возвращает прибыль для заданного количества ночей "nights" в рамках одного сезона "season"
    для номера со стоимостью "per_night_price" в сутки и стоимостью обслуживания "per_service_price"
    '''
    money_in = nights * (season['multiplier'] * per_night_price)
    money_out = get_number_of_services_by_nights_count(nights) * (season['multiplier'] * per_service_price)
    return money_in - money_out

def calculate_profit(row):
    '''Подсчитывает прибыль для бронирования, информация о бронировании поступает
    в функцию в объекте Series "row"
    '''
    per_night_price = row['per_night_price']
    per_service_price = row['per_service_price']
    is_canceled = row['is_canceled']

    if 'is_canceled_predicted' not in row:
        # Series без предсказаний об отмене бронирования (столбец `is_canceled_predicted`)
        # В этом случае, если клиент отменил бронь, то мы не получаем ничего
        if is_canceled:
            return 0
    else:
        # Series с предсказаниями об отмене бронирования (столбец `is_canceled_predicted`)
        is_canceled_predicted = row['is_canceled_predicted']
        if is_canceled and not is_canceled_predicted:
            # Клиент отменил бронирование, и мы не предсказали, что он это может сделать
            # В этом случае мы опять же не получаем ничего
            return 0
        elif is_canceled and is_canceled_predicted:
            # Клиент отменил бронирование, но мы это предсказали
            # Тогда снимаем с него 80% от аренды за одни сутки и одно обслуживание номера
            income = (0.8 * per_night_price) + (0.8 * per_service_price)
            return income
        

    in_month = row['in_date'].month
    out_month = row['out_date'].month

    in_season = SEASONS_DICT[in_month]
    out_season = SEASONS_DICT[out_month]

    if in_season['season_id'] == out_season['season_id']:
        # Нет пересечения сезонов в рамках данного бронирования
        income = calculate_partial_income(row['total_nights'], in_season, per_night_price, per_service_price)
    else:
        # При бронировании есть переход из одного сезона в другой, будем считать их отдельно как левый-правый:
        # - "left..."  - относится к сезону, ближайшему к НАЧАЛУ бронирования
        # - "right..." - относятся к сезону, ближайшему к ОКОНЧАНИЮ бронирования
        left_next_month = row['in_date'] + pd.DateOffset(months=1) - pd.offsets.MonthBegin(1)
        left_nights = (left_next_month - row['in_date']).days
        left_income = calculate_partial_income(left_nights, in_season, per_night_price, per_service_price)
        
        right_curr_month = row['out_date'] + pd.offsets.MonthBegin(-1)
        right_nights = (row['out_date'] - right_curr_month).days
        right_income = calculate_partial_income(right_nights, out_season, per_night_price, per_service_price)

        income = left_income + right_income

    return income

# Протестируем функцию на некоторых данных
test_in_date = '2015-08-28'
test_out_date = '2015-09-03'
test_total_nights = 6
ser = pd.Series({'in_date': pd.to_datetime(test_in_date),
                 'out_date': pd.to_datetime(test_out_date),
                 'total_nights': test_total_nights,
                 'id': 777,
                 'per_service_price': 450,
                 'per_night_price': 1000,
                 'is_canceled': False})

calculate_profit(ser)
````

Теперь применим нашу функцию к тестовой выборке, чтобы сохранить информацию о прибыли для каждого из бронирований в новом столбце `initial_profit`. Если бронирование было отменено, то значением данного столбца будет `0`:

````python
df_test['initial_profit'] = df_test.apply(calculate_profit, axis=1)
display(df_test.sample(n=5, random_state=1))
````

Наконец, можем получить информацию об изначальной  прибыли на тестовой выборке, просуммировав все значения в столбце `initial_profit`:

````python
def get_money_pretty(money):
    return ('{:,}'.format(money).replace(',', ' '))

initial_profit = df_test['initial_profit'].sum()
print(f'Прибыль отеля без внедрения депозитов: {get_money_pretty(initial_profit)}')
````

### Шаг 4. Разработка модели ML

- `customer_type` - тип заказчика;
- `distribution_channel` - канал дистрибуции заказа;
- `meal` - опции заказа;
- `reserved_room_type` - тариф (тип комнаты)

Что касается месяца заезда (`month`), то пока воздержимся от его кодирования, поскольку у меня есть сомнения в его использовании про обучении моделей - есть подозрения, что неделя заезда (`arrival_date_week_number`) может представлять ту же самую информацию для процесса обучения.

Для кодирования будем использовать класс `OneHotEncoder`, который **обучим только на тренировочных данных**, а применим его после обучени дважды - на тренировочном и на тестовом наборе. Аргументация в пользу использования такого подхода заключается в том, что если в тестовом наборе окажутся _новые_ (отсутствующие в тренировочном наборе) значения кодированных категорий, то они получат значение `0`. Это в какой-то степени моделирует возможную ситуацию реального мира: например, когда в новых данных, поступающих в обученную модель, может оказаться какая-то редкая страна бронирующего, которая не встречалась в процессе обучения и тестирования модели. И модель должна быть способна справиться с такой ситуацией, то есть корректно обработать случаи с появлением не встречавшихся ранее значений кодированных категорий.

````python
columns_to_encode = [
        'distribution_channel',
        'country',
        'customer_type',
        'meal',
        'reserved_room_type'
    ]

# Создаем объект OneHotEncoder, который впоследствии будем использовать дважды
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

# ... и обучаем его и прогоняем на тренировочной выборке
df_train_transformed_cols = one_hot_encoder.fit_transform(df_train[columns_to_encode])

# Получаем имена категорий - для корректного именования столбцов
feature_names = one_hot_encoder.get_feature_names(columns_to_encode)
    
def do_one_hot_encoding(df, df_transformed_cols, feature_names):
    # Создаём DataFrame на основе кодированных категорий
    df_encoded_cols = pd.DataFrame(df_transformed_cols, index=df.index, columns=feature_names)

    # В ещё одном DataFrame будем хранить все прочие (не кодированные) категории
    df_other_cols = df.drop(columns=columns_to_encode)

    # Наконец, сольём упомянутые выше DataFrame'ы между собой
    df_encd = pd.concat([df_encoded_cols, df_other_cols], axis=1)
    
    return df_encd

# Прогоняем кодировщик на тренировочной выборке
df_train_encd = do_one_hot_encoding(df_train, df_train_transformed_cols, feature_names)

# ... и протестируем результат на каком-нибудь входе
display(df_train_encd[df_train['id'] == 12314][['id', 'lead_time', 'meal_BB', 'country_PRT', 'country_ITA']])
````

Теперь применим наш обученный объект `one_hot_encoder` к тестовому набору:

````python
df_test_transformed_cols = one_hot_encoder.transform(df_test[columns_to_encode])

# Теперь прогоняем кодировщик на тестовой выборке
df_test_encd = do_one_hot_encoding(df_test, df_test_transformed_cols, feature_names)

# ... и снова протестируем результат на каком-нибудь входе
display(df_test_encd[df_test['id'] == 6086][['id', 'lead_time', 'meal_BB', 'country_PRT', 'country_ITA']])
````

````python
excess_features_train = {
    'id', 'arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month', 
    'per_night_price', 'per_service_price', 'in_date', 'out_date', 'is_canceled' }

excess_features_test = excess_features_train.copy()
excess_features_test.add('initial_profit')

all_features = set(df_train_encd.columns)

features_train = df_train_encd.drop(columns=excess_features_train, axis=1)
features_test = df_test_encd.drop(columns=excess_features_test, axis=1)

target_train = df_train_encd['is_canceled']
target_test = df_test_encd['is_canceled']

# Проверим размеры получившихся наборов
print(features_train.shape)
print(target_train.shape)
print()
print(features_test.shape)
print(target_test.shape)

# И сами наборы (частично)
display(features_train.head())
display(target_train.to_frame().head())
````

Нам также нужно выбрать метрику для обучения, поскольку "_верность_" ("_accuracy_") может оказаться неподходящей для нашего случая с дисбалансом классов. Поэтому будем использовать **F1-меру в качестве нашей метрики обучения**.

Для выбора наилучшей модели будем рассматривать _логистическую регрессию_, _дерево решений_ и _случайный лес_ с последующим выбором наилучшей из них с точки зрения нашей метрики качества (F1-меры).

Подготовим словари, которые будут задавать настраиваемые гиперпараметры каждой из этих трёх моделей:

````python
logreg_params = dict(C=[200, 175, 150, 125, 100, 50, 30, 20, 10, 5, 3.0, 2.0,
                        1.0, 0.75, 0.5, 0.25, 0.1, 0.05, 0.025, 0.01])

tree_params = dict(max_depth=range(3, 150),
                   min_samples_split=range(10, 250),
                   min_samples_leaf=range(10, 250),
                   max_features=['auto', 'sqrt'])

forest_params = tree_params.copy()
forest_params['n_estimators'] = range(10, 500)
forest_params['bootstrap'] = [True, False]
````

Теперь попробуем обучить несколько моделей на обучающей выборке для того, чтобы затем выбрать среди них лучшую. Для перебора возможных комбинаций гиперпараметров будем использовать **рандомизированный поиск с кросс-валидацией** (класс `RandomizedSearchCV`), а в качестве оценки качества модели, как было упомянуто выше, будем использовать F1-меру.

Создадим вспомогательную функцию для обучения моделей и подбора лучших значений гиперпараметров, затем сохраним наилучшие с точки зрения F1-меры модели в переменные `best_logreg_model`, `best_tree_model` и `best_forest_model`, а также прогоним её с разными моделями.

Для устранения негативного эффекта от дисбаланса классов будем делать **взвешивание классов**, используя аргумент `class_weight='balanced'` для всех создаваемых нами моделей:

````python
def get_best_model_by_hyperparams_search(name, model, features_used, target_used, params, scoring, iters):
    random_search = RandomizedSearchCV(model, params, scoring=scoring, cv=10, n_iter=iters, random_state=12345)
    random_search.fit(features_used, target_used)
    
    print()
    print(f'Результаты для модели "{name}":\n')
    print(f'\t- метрика качества :\t{random_search.best_score_}\n')
    print(f'\t- гиперпараметры   :\t{random_search.best_params_}\n')
    
    return (random_search.best_estimator_, random_search.best_score_)

best_weighted_logreg_model, best_weighted_logreg_score = get_best_model_by_hyperparams_search(
        'Logistic Regression (взвешенная)',
        LogisticRegression(class_weight='balanced', random_state=12345),
        features_train,
        target_train,
        logreg_params,
        'f1',
        10)

best_weighted_tree_model, best_weighted_tree_score = get_best_model_by_hyperparams_search(
        'Decision Tree Classifier (взвешенное)',
        DecisionTreeClassifier(class_weight='balanced', random_state=12345),
        features_train,
        target_train,
        tree_params,
        'f1',
        1000)

best_weighted_forest_model, best_weighted_forest_score = get_best_model_by_hyperparams_search(
        'Random Forest Classifier (взвешенный)',
        RandomForestClassifier(class_weight='balanced', random_state=12345),
        features_train,
        target_train,
        forest_params,
        'f1',
        1000)
````

Вместе с тем, пока модель разрабатывается и встраивается в процесс компании, можно незамедлительно использовать определённые наработки, характеризующие типичного "ненадёжного клиента" - чтобы начать вводить депозиты в практику ещё до окончательной разработки модели и получать от этого дополнительную финансовую выгоду. Так, простейшей рекомендацией может быть требование внесения депозита (предоплаты) для всех клиентов из Португалии - это можно начать делать буквально "_прямо сейчас_", пока никакой модели ещё не разработано, или она ещё не внедрена в эксплуатацию.

В качестве дальнейшей работы по улучшению разработанной модели здесь мог бы быть учёт ошибок `I` и `II` рода в зависимости от того, какая ошибка более важна для бизнеса: либо "_ложная тревога_" и снятие депозита с надёжного клиента, либо "_упущения в учёте_" и бездепозитное размещение бронирования ненадёжным клиентом.
