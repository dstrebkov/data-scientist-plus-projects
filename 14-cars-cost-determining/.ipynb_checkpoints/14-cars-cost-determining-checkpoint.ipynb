## Обработка дублированных анкет

Сначала поищем явные дубли:

```python
print(f'Количество явных дублей: {data.duplicated().sum()}')
print()
display(data[data.duplicated()])
```

Удаляем их и проверяем, что они исчезли:

```python
data = data.drop_duplicates()
print(f'Количество явных дублей: {data.duplicated().sum()}')
```

Теперь попробуем поискать дубли более тщательно. Необходимость этого связана с наличием полей с датами (`DateCrawled`, `DateCreated` и `LastSeen`), а конкретнее с возможным появлением неявных дублей, т.е. дублей, которые отличаются только содержимым одного или нескольких из этих полей.

Гипотезы о том, почему каждый из этих столбцов в теории может способствовать появлению неявных дублей, приведены ниже:

- столбец `DateCrawled` — если вдруг одна и та же анкета была скачана дважды в разные моменты времени, то она будет неявно продублирована;

- столбец `LastSeen` — если вдруг автор анкеты был активен в процессе краулинга, то обновление этого столбца может сделать соответствующую анкету уникальной.

```python
columns_subset = set(data.columns)
columns_subset.remove('DateCrawled')
columns_subset.remove('LastSeen')
columns_subset.remove('DateCreated')

display(data[columns_subset])

print(f'Количество неявных дублей: {data[columns_subset].duplicated().sum()}')
```

```python
index_duplicated = data[columns_subset][data[columns_subset].duplicated()].index
print(index_duplicated)
print()
data = data.drop(index=index_duplicated)
display(data)
```

Убеждаемся, что неявных дублей в наборе данных больше не осталось:

```python
print(f'Количество неявных дублей: {data[columns_subset].duplicated().sum()}')
```

#### <b>Анализ признака `VehicleType`</b>

```python
def get_group_agg_count_for_feature(feature):
    res_df = (data[feature].value_counts(dropna=False)
                  .to_frame()
                  .reset_index()
                  .fillna('NaN')
                  .groupby('index')
                  .agg('sum')
                  .reset_index()
                  .rename(columns={'index': feature, feature: 'Count'}))
    return res_df

vehicle_type = get_group_agg_count_for_feature('VehicleType')
display(vehicle_type)

fig, ax = plt.subplots(figsize=(22, 6))
_ = sns.barplot(data=vehicle_type, x='VehicleType', y='Count', ax=ax)
_ = ax.set_xlabel('Тип кузова')
_ = ax.set_ylabel('Количество анкет')
_ = ax.set_title('Столбчатая диаграмма по типам кузовов')
_ = ax.grid(True)
```

Судя по выдаче, здесь отсутствующим значением можно считать только явные `np.NaN`, таким образом в качестве аномалий здесь есть только явные пропуски:

## Очистка (удаление) анкет
Начнём с очистки (удаления анкет согласно решениям, сделанным выше):

```python
# 1. Соберём все анкеты с ценой ниже €100
rm_index = data.query('Price < 100').index

# 2. Соберём все анкеты с аномальным годом регистрации
rm_index = rm_index.union(registration_year_df.index)

# 3. Соберём автомобили с мощностью от 0 до 10 и с мощностью больше 1000 
rm_index = rm_index.union(data.query('Power > 0 and Power < 10').index)
rm_index = rm_index.union(data.query('Power > 1000').index)

# Удалим собранные анкеты по индексу
data.drop(rm_index, inplace=True)

# Удаляем столбцы 'DateCrawled' и 'NumberOfPictures' как бесполезные
data.drop('DateCrawled', inplace=True, axis=1)
data.drop('NumberOfPictures', inplace=True, axis=1)

# Заменяем 0 на NaN для мощности и месяца регистрации
data['Power'] = data['Power'].replace(0, np.nan)
data['RegistrationMonth'] = data['RegistrationMonth'].replace(0, np.nan)

display(data)
```

Для определения способа заполнения пропусков можно остановиться на категории с типом кузова `VehicleType`.

Первое, что приходит здесь в голову, что нельзя заполнить пропуски типа кузова просто выбрав, скажем, моду распределения. Так, если модой окажется тип кузова _"седан"_ и мы выставим его для всех пропусков, то не исключено появление несуществующих комбинаций авто с типом кузова. Например, если бы у нас оказалась анкета автомобиля _"Лада Нива"_ с пропущенным типом кузова, то замена пропуска на _"седан"_ привела бы к появлению недопустимой комбинации признаков, ведь автомобиля _"Лада Нива"_ с кузовом _"седан"_ не существует.

Возможно, мы могли бы проставить для каждой модели (если это значение непусто) автомобиля наиболее часто встречающийся тип кузова. Однако, в случае автомобиля _"Фольксваген Гольф"_ можно заметить, что значения _"small"_ и _"wagon"_ хоть и уступают наиболее частотному значению _"sedan"_, но всё равно встречаются тоже сравнительно часто. Не исключено, что с другими автомобилями ситуация может быть ещё более неоднозначной.

Вероятно, мы могли бы более точно восстанавливать пропуски в столбце `VehicleType`, если бы учитывали остальные признаки. В частности, помимо бренда и модели, здесь нам могли бы оказаться полезными:

- год регистрации автомобиля (`RegistrationYear`) — возможно, что некоторые типы кузова автомобиля появились в определённый год в модельном ряду;

- цена (`Price`) — возможно, что автомобили с определённым типом кузова различаются по своей цене (скажем, седан дороже, чем хетчбэк, или наоборот);

Исходя из этого, попробуем заполнить пропуски, находя ближайшего соседа (либо нескольких ближайших соседей) в пространстве признаков анкеты. Для реализации данного способа выпишем отдельно категориальные и численные признаки, которые содержат пропуски в настоящий момент.

Категориальные:
- `VehicleType`
- `Gearbox`
- `Model`
- `FuelType`
- `NotRepaired`
- `RegistrationMonth`

Численные:
- `Power`

В случае числового признака, мы могли бы брать несколько ближайших соседей и брать среднее среди соседей для заполнения пропуска, используя `KNNImputer`. В случае категориальных признаков мы должны будем ориентироваться на единственного ближайшего соседа, поскольку понятия среднего для категориальных данных нет. Кроме того, нам потребуется кодировать категориальные признаки в числовое представление через `LabelEncoder`, сохраняя при этом имеющиеся пропуску `np.NaN` нетронутыми.

```python
#####################################################################################
# Заполнение пропусков
#####################################################################################
COLS_TO_ENCODE = {
    'Brand'            : LabelEncoder(),
    'VehicleType'      : LabelEncoder(),
    'Gearbox'          : LabelEncoder(),
    'Model'            : LabelEncoder(),
    'FuelType'         : LabelEncoder(),
    'NotRepaired'      : LabelEncoder(),
    'RegistrationMonth': LabelEncoder(),
}

IMPUTATIONS = [
    # Для заполнения пропусков в столбцах 'VehicleType' и 'Model'
    {
        'imputer': KNNImputer(n_neighbors=1),
        'cols': ['Model', 'Brand', 'VehicleType']
    },
    # Для заполнения пропусков в остальных категориальных столбцах
    {
        'imputer': KNNImputer(n_neighbors=1),
        'cols': ['VehicleType', 'RegistrationYear', 'Gearbox', 'Model', 'RegistrationMonth', 'FuelType', 'Brand', 'NotRepaired', 'PostalCode']
    },
    # Для заполнения пропусков в столбцах 'Power'
    {
        'imputer': KNNImputer(n_neighbors=10, weights='distance'),
        'cols': ['Price', 'Power', 'Kilometer', 'Brand', 'Model', 'RegistrationYear']
    }
]

# Из-за проблем с памятью (?) при прогоне всего набора данных через Imputer, поделим наш набор данных на IMPUTE_SPLITS_COUNT частей,
# и каждую из частей будем обрабатывать объектом Imputer по отдельности
IMPUTE_SPLITS_COUNT = 10

TARGET_ENCODER = TargetEncoder(smoothing=1.1, min_samples_leaf=2,
                               cols=['VehicleType', 'Gearbox', 'Model', 'FuelType', 'Brand', 'NotRepaired', 'PostalCode'])
SCALER = StandardScaler()

def encode_labels_keeping_nans(df, cols_to_encode_dict, is_training):
    for col, enc in cols_to_encode_dict.items():
        # Кодируем сохраняя NaN'ы
        if is_training:
            df[col] = pd.Series(data=enc.fit_transform(df[col][df[col].notnull()]), index=df[col][df[col].notnull()].index)
        else:
            df[col] = pd.Series(data=enc.transform(df[col][df[col].notnull()]), index=df[col][df[col].notnull()].index)
    return df
        
def split_for_impution(df, impute_splits_count):
    impute_splits_size = df.shape[0] // impute_splits_count
    impute_splits_list = []
    for i in range(impute_splits_count):
        start = i * impute_splits_size
        end = df.shape[0] if i+1 == impute_splits_count else (i+1) * impute_splits_size
        impute_split = df.iloc[start: end].index
        impute_splits_list.append(impute_split)
    return impute_splits_list

def impute_nans(df, impute_splits, imputations, is_training):
    for impute_split in impute_splits:
        for imputation in imputations:
            imputer = imputation['imputer']
            cols = imputation['cols']
            if is_training:
                df.loc[impute_split, cols] = imputer.fit_transform(df.loc[impute_split, cols])
            else:
                df.loc[impute_split, cols] = imputer.transform(df.loc[impute_split, cols])
    return df

def inverse_encode_labels(df, cols_to_encode_dict):
    for col, enc in cols_to_encode_dict.items():
        df[col] = enc.inverse_transform(df[col].astype(int))
    return df

def make_imputation(df, imputations, is_training):
    df = encode_labels_keeping_nans(df, COLS_TO_ENCODE, is_training)
    impute_splits = split_for_impution(df, IMPUTE_SPLITS_COUNT)
    df = impute_nans(df, impute_splits, imputations, is_training)
    df = inverse_encode_labels(df, COLS_TO_ENCODE)
    df['PostalCode'] = df['PostalCode'].astype(int)
    return df

#####################################################################################
# Объединение месяца и года регистрации в один признак "Registration"
#####################################################################################
def join_registration_month_year(df):
    df['RegistrationYear'] = df['RegistrationYear'].astype(int)
    df['RegistrationMonth'] = df['RegistrationMonth'].astype(int)
    df['Registration'] = df['RegistrationMonth'].astype(str) + '-' + df['RegistrationYear'].astype(str)
    df['Registration'] = pd.to_datetime(df['Registration'], format='%m-%Y')
    df = df.drop('RegistrationMonth', axis=1)
    df = df.drop('RegistrationYear', axis=1)
    return df

def make_dates_ordinal(df):
    df['DateCreated'] = pd.to_datetime(df['DateCreated'], format='%Y-%m-%d %H:%M:%S')
    df['LastSeen'] = pd.to_datetime(df['LastSeen'], format='%Y-%m-%d %H:%M:%S')
    df['DateCreated'] = df['DateCreated'].apply(lambda x: x.toordinal())
    df['LastSeen'] = df['LastSeen'].apply(lambda x: x.toordinal())
    df['Registration'] = df['Registration'].apply(lambda x: x.toordinal())
    return df

def encode_dates(df):
    df = join_registration_month_year(df)
    df = make_dates_ordinal(df)
    return df

#####################################################################################
# Кодирование категориальных переменных
#####################################################################################
# Поскольку One-Hot кодирование может привести к очень большому увеличению количества признаков (категории `Model` и `Brand` имеют много значений),
# попробуем сделать кодирование с помощью техники Target Encoding
def encode_categorical_features(df, enc, is_training):
    if is_training:
        enc.fit(df, df['Price'])
    return enc.transform(df, df['Price'])

#####################################################################################
# Шкалирование числовых признаков
#####################################################################################
def scale_numeric_features(X, scaler, is_training):
    if is_training:
        scaler = scaler.fit(X)
    X_scaled = scaler.transform(X)
    return X_scaled
```

## Подготовка признаков обучающего набора (прогон)

```python
display(Xy_train.head())

Xy_train = make_imputation(Xy_train, IMPUTATIONS, is_training=True)
Xy_valid = make_imputation(Xy_valid, IMPUTATIONS, is_training=False)
Xy_test  = make_imputation(Xy_test, IMPUTATIONS, is_training=False)
display(Xy_train.head())

Xy_train = encode_dates(Xy_train)
Xy_valid = encode_dates(Xy_valid)
Xy_test  = encode_dates(Xy_test)
display(Xy_train.head())

Xy_train = encode_categorical_features(Xy_train, TARGET_ENCODER, is_training=True)
Xy_valid = encode_categorical_features(Xy_valid, TARGET_ENCODER, is_training=False)
Xy_test  = encode_categorical_features(Xy_test, TARGET_ENCODER, is_training=False)
display(Xy_train.head())

X_train = Xy_train.drop('Price', axis=1)
X_valid = Xy_valid.drop('Price', axis=1)
X_test  = Xy_test.drop('Price', axis=1)
y_train = Xy_train['Price']
y_valid = Xy_valid['Price']
y_test  = Xy_test['Price']
X_train = scale_numeric_features(X_train, SCALER, is_training=True)
X_valid = scale_numeric_features(X_valid, SCALER, is_training=False)
X_test  = scale_numeric_features(X_test, SCALER, is_training=False)

X_train_valid = np.concatenate([X_train, X_valid], axis=0)
y_train_valid = np.concatenate([y_train, y_valid], axis=0)

print(f'X_train.shape={X_train.shape}, y_train.shape={y_train.shape}')
print(f'X_valid.shape={X_valid.shape}, y_valid.shape={y_valid.shape}')
print(f'X_test.shape={X_test.shape}, y_test.shape={y_test.shape}')
print(f'X_train_valid.shape={X_train_valid.shape}, y_train_valid.shape={y_train_valid.shape}')
```

## Обучение линейных моделей

Для начала создадим несколько вспомогательных функций:

```python
def search_best_hyperparams(model, params):
    grid_search = RandomizedSearchCV(estimator=model,
                                     param_distributions=params,
                                     scoring='neg_root_mean_squared_error',
                                     cv=10,
                                     n_jobs=-1,
                                     n_iter=3,
                                     random_state=0)
    grid_search = grid_search.fit(X_train, y_train)
    best_score = -1 * grid_search.best_score_
    best_params = grid_search.best_params_
    print(f'Hyperparameter tuning for "{model.__class__.__name__}":\n\tBest RMSE on Train = {best_score}\n\tBest params = {best_params}')
    return best_score, best_params

def calculate_rmse(model, use_test_set=False):
    print(f'Model "{model.__class__.__name__}" results:')
    if not use_test_set:
        train_rmse = mean_squared_error(y_train, model.predict(X_train), squared=False)
        valid_rmse = mean_squared_error(y_valid, model.predict(X_valid), squared=False)
        result = f"\tRMSE on Train = {train_rmse}\n\tRMSE on Valid = {valid_rmse}"
    else:
        # Переучиваемся на объединённом наборе из тренировочного и валидационного
        model = model.fit(X_train_valid, y_train_valid)
        test_rmse = mean_squared_error(y_test, model.predict(X_test), squared=False)
        result = f"\tRMSE on Test  = {test_rmse}"
    return result
```

Теперь обучим 3 модели линейной регрессии: без регуляризации, с L1-регуляризацией и с L2-регуляризацией. Для двух последних моделей подбор наилучшего гиперпараметра `alpha`, определяющего степень регуляризации, определим рандомизированным поиском с кросс-валидацией на тренировочном наборе.

```python
%%time

def train_linear_regression():
    linear_reg = LinearRegression(n_jobs=-1)
    linear_reg = linear_reg.fit(X_train, y_train)
    print(calculate_rmse(linear_reg))

    linear_model_params = [{'alpha': np.linspace(0, 5, 101)}]

    lasso_best_score, lasso_best_params = search_best_hyperparams(Lasso(random_state=0), linear_model_params)
    ridge_best_score, ridge_best_params = search_best_hyperparams(Ridge(random_state=0), linear_model_params)

    tuned_lasso = Lasso(alpha=lasso_best_params['alpha'])
    tuned_lasso.fit(X_train, y_train)
    print(calculate_rmse(tuned_lasso))

    tuned_ridge = Ridge(alpha=ridge_best_params['alpha'])
    tuned_ridge.fit(X_train, y_train)
    print(calculate_rmse(tuned_ridge))
    return linear_reg

linear_reg = train_linear_regression()
```

Видим, что ни одна из моделей (в том числе регуляризованные варианты) не оказалось способна добиться целевого значения по RMSE  метрике, равного `€2500`.

Теперь попробуем получить информацию о важности признаков исходя из значений коэффициентов обученной линейной регрессии:

```python
coefs = pd.DataFrame(linear_reg.coef_, columns=['Coefficients'], index=Xy_train.drop('Price', axis=1).columns)
display(coefs)
fig, ax = plt.subplots(figsize=(18, 4))
_ = coefs.plot(kind='bar', ax=ax)
_ = ax.grid()
```

Глядя на выдачу выше, можно заметить наиболее важных призника с точки зрения их значения по модулю:

- `Power`
- `Model`
- `Kilometer`
- `Registration`

Можно также отметить отрицательное значение коэффициента при признаке `Kilometer`, что кажется логичным: при увеличении пробега уменьшается стоимость автомобиля.

## Обучение LightGBM

```python
%%time

def train_lgbm(is_tuning):
    lgb_train = lgb.Dataset(X_train, y_train)
    lgb_eval = lgb.Dataset(X_valid, y_valid, reference=lgb_train)

    if is_tuning:
        lgbm_params = {
            'num_leaves': [7, 14, 21, 28, 31, 50],
            'learning_rate': [0.1, 0.03, 0.003, 0.03, 0.1, 0.15, 0.25, 0.3],
            'max_depth': [-1, 3, 5],
            'n_estimators': [50, 100, 200, 500, 700],
        }

        lgbm_best_score, lgbm_best_params = search_best_hyperparams(lgb.LGBMRegressor(random_state=0, n_jobs=-1), lgbm_params)
        tuned_lgbm = lgb.LGBMRegressor(random_state=0,
                                       n_jobs=-1,
                                       n_estimators=lgbm_best_params['n_estimators'],
                                       num_leaves=lgbm_best_params['num_leaves'],
                                       learning_rate=lgbm_best_params['learning_rate'],
                                       max_depth=lgbm_best_params['max_depth'])
        tuned_lgbm.fit(X_train, y_train)
        print(calculate_rmse(tuned_lgbm))
        print(tuned_lgbm)
    else:
        tuned_lgbm = lgb.LGBMRegressor(random_state=0,
                                       n_jobs=-1,
                                       n_estimators=500,
                                       num_leaves=50,
                                       learning_rate=0.15,
                                       max_depth=-1)
        tuned_lgbm.fit(X_train, y_train)
        print(calculate_rmse(tuned_lgbm))
    return tuned_lgbm

tuned_lgbm = train_lgbm(is_tuning=False)
```

## Обучение CatBoost

```python
%%time

def train_catboost(is_tuning):
    if is_tuning:
        catboost_params = {
              'iterations': [50, 75, 100, 150, 200],
              'depth': [4, 5, 6, 7, 8],
              'learning_rate': [0.001, 0.01, 0.1, 0.15, 0.2, 0.25, 0.3, 0.5],
             }

        catboost_best_score, catboost_best_params = search_best_hyperparams(CatBoostRegressor(random_state=0, silent=True), catboost_params)
        tuned_catboost = CatBoostRegressor(random_state=0,
                                           iterations=catboost_best_params['iterations'],
                                           depth=catboost_best_params['depth'],
                                           learning_rate=catboost_best_params['learning_rate'])
        tuned_catboost.fit(X_train, y_train)
        print(calculate_rmse(tuned_catboost))
        print(tuned_catboost)
    else:
        tuned_catboost = CatBoostRegressor(random_state=0,
                                           iterations=200,
                                           depth=7,
                                           learning_rate=0.5,
                                           silent=True)
        tuned_catboost.fit(X_train, y_train)
        print(calculate_rmse(tuned_catboost))
    return tuned_catboost

tuned_catboost = train_catboost(is_tuning=False)
```

```python
X_train = Xy_train[['Power', 'Model', 'Kilometer', 'Registration']]
X_valid = Xy_valid[['Power', 'Model', 'Kilometer', 'Registration']]
X_test  = Xy_test[['Power', 'Model', 'Kilometer', 'Registration']]
y_train = Xy_train['Price']
y_valid = Xy_valid['Price']
y_test  = Xy_test['Price']
X_train = scale_numeric_features(X_train, SCALER, is_training=True)
X_valid = scale_numeric_features(X_valid, SCALER, is_training=False)
X_test  = scale_numeric_features(X_test, SCALER, is_training=False)

X_train_valid = np.concatenate([X_train, X_valid], axis=0)
y_train_valid = np.concatenate([y_train, y_valid], axis=0)

print(f'X_train.shape={X_train.shape}, y_train.shape={y_train.shape}')
print(f'X_valid.shape={X_valid.shape}, y_valid.shape={y_valid.shape}')
print(f'X_test.shape={X_test.shape}, y_test.shape={y_test.shape}')
print(f'X_train_valid.shape={X_train_valid.shape}, y_train_valid.shape={y_train_valid.shape}')
print()

def train_random_forest(is_tuning):
    if is_tuning:
        forest_params = [{
            'n_estimators': range(2, 200),
            'max_depth': range(2, 300),
            'min_samples_split': range(1, 100),
            'min_samples_leaf': range(1, 100),
            'max_features': range(1, 12)
        }]

        forest_best_score, forest_best_params = search_best_hyperparams(RandomForestRegressor(random_state=0, n_jobs=-1), forest_params)
        tuned_forest = RandomForestRegressor(random_state=0,
                                             n_jobs=-1,
                                             n_estimators=forest_best_params['n_estimators'],
                                             max_depth=forest_best_params['max_depth'],
                                             min_samples_split=forest_best_params['min_samples_split'],
                                             min_samples_leaf=forest_best_params['min_samples_leaf'],
                                             max_features=forest_best_params['max_features'])
        tuned_forest.fit(X_train, y_train)
        print(calculate_rmse(tuned_forest))
        print(tuned_forest)
    else:
        tuned_forest = RandomForestRegressor(random_state=0,
                                             n_jobs=-1,
                                             n_estimators=30,
                                             max_depth=5,
                                             min_samples_split=10,
                                             min_samples_leaf=5,
                                             max_features=12)
        tuned_forest.fit(X_train, y_train)
        print(calculate_rmse(tuned_forest))
    return tuned_forest
        
tuned_forest = train_random_forest(is_tuning=False)

tuned_linear_reg = train_linear_regression()
tuned_lgbm = train_lgbm(is_tuning=False)
tuned_catboost = train_catboost(is_tuning=False)

print()
print(calculate_rmse(linear_reg, use_test_set=True))
print(calculate_rmse(tuned_forest, use_test_set=True))
print(calculate_rmse(tuned_lgbm, use_test_set=True))
print(calculate_rmse(tuned_catboost, use_test_set=True))
```

## Выводы

Прежде всего отметим, что даже после объединения тренировочного и валидационного набора модель линейной регрессии (в том числе её регуляризированные варианты) оказалась не способна добиться целевого показателя в `€2500` по RMSE метрике на тестовом наборе, продемонстрировав качество `€2813`.

Модели градиентного бустинга показали значительно более высокое качество, в частности LightGBM показала качество `€1549` на тестовом наборе, а CatBoost — `€1602`.

Также был проведён эксперимент по использованию усечённого набора из 4 наиболее значимых (с точки зрения коэффициентов линейной регрессии) признаков: `Power`, `Model`, `Kilometer`, `Registration`. Качество всех опробованных на таком наборе моделей ухудшилось, но не очень значительно:

- LightGBM ухудшила RMSE с `€1549` до `€1699`;
- CatBoost ухудшила RMSE с `€1602` до `€1757`;
- линейная регрессия ухудшила RMSE с `€2813` до `€2953`.

Отдельно отметим, что в случае усечённого набора признаков была успешно обучена модель RandomForestRegressor, которая после настройки гиперпараметров оказалась способна добиться качества чуть выше требуемого: `€2336`. Тем не менее, учитывая сложность её обучения (с точки зрения ресурсов компьютера), а также тот факт, что вторым после качества предсказания приоритетом для заказчика является время обучения модели, можно в качестве итоговой рекомендации советовать <u>использовать модель градиентного бустинга</u>, LightGBM или CatBoost. Кроме того (если я не ошибаюсь), данные библиотеки способны работать с категориальными признаками напрямую, а значит можно не осуществлять их кодирование, сократив тем самым время на предобработку.
